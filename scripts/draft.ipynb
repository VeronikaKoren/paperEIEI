```python
######################
#####IMPORT TOOLS####
import scipy
import os
from os import path
sys.path.append("./lib/netGraphClass/");
sys.path.append("./lib/IFNetworkClass/");
sys.path.append("./lib/rateNetworkClass/");
sys.path.append("./lib/matClass/");
sys.path.append("./lib/daftClass/");
sys.path.append("./lib/toolsClass/");
import IFNetworkClass 
reload(IFNetworkClass)
import rateNetworkClass 
reload(rateNetworkClass)
import matClass
reload(matClass)
import daftClass
reload(daftClass)
import netGraphClass
reload(netGraphClass)
import toolsClass
reload(toolsClass)
np.set_printoptions(precision=4,suppress=True)

######################
#####I MODEL SET#####
nPop=1;
#M=1;typeExp=array(["variable" for i in xrange(M)]);typeA="leak";typeCon="fromGamma";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=False;
#M=1;typeExp=array(["constant" for i in xrange(M)]);typeA="leak";typeCon="fromGamma";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=True;
#M=1;typeExp=array(["constant" for i in xrange(M)]);typeA="leak";typeCon="fromOmega";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=False;
#M=1;typeExp=array(["steps" for i in xrange(M)]);typeA="leak";typeCon="fromGamma";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=False;
#M=1;typeExp=array(["steps" for i in xrange(M)]);typeA="integrator";typeCon="fromGamma";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=False;
#M=1;typeExp=array(["constant" for i in xrange(M)]);typeA="leak";typeCon="fromGamma";typeStru="dense";typeDelay="off";typeReset="inOmega";filterSpikes=True;
#M=100;typeExp=array(["variable" for i in xrange(M)]);typeA="leak";typeCon="fromGamma";typeStru="sparse";typeDelay="off";typeReset="inOmega";filterSpikes=True;
M=3;typeExp=array(["constant" for i in xrange(M)]);typeA="leak";typeCon="fromOmega";typeStru="sparse";typeDelay="off";typeReset="explicit";filterSpikes=True;

############################
#####INIT OBJECTS##########
print("");
print("CAN BUILD NET AND RATE OBJECT");
myRate=rateNetworkClass.rateNetworkClass(_nPop=nPop,_M=M);
myNet=IFNetworkClass.IFNetworkClass(_nPop=nPop,_M=M);
myNet.printControl=False;

############################
#####SET COST OPTIONS########
if typeReset=="inOmega":
    myNet.nu=0.; #linear cost
    myNet.mu=0.; #quadratic cost
elif typeReset=="explicit":
    myNet.nu=0.; #linear cost
    myNet.mu=10.; #quadratic cost

##############################
#####SET NUMBER OF NEURONS####
if typeCon=="fromGamma":
    myNet.set_N(0,3);
if typeCon=="fromOmega":
    myNet.set_N(0,M);
    
##############################
#####SET COMMANDS FOR POP#####
myNet.neuPop[0].isCommanded=True;
        
######################
######SET RUN/CONTROL OPTIONS##
runTime=1000.;dt=0.1;
myRate.runTime=runTime;myNet.runTime=runTime;
myRate.set_simulationClock(_simuClock=dt);myNet.set_simulationClock(_simuClock=dt);
myNet.filterSpikes=filterSpikes;
    
##############################
###BUILD THE INPUTS ###
if typeExp[i]=="constant":
    for pop in xrange(nPop):
        myRate.ratePop[pop].sigmaV=0;
        myNet.neuPop[pop].sigmaV=0.1;
        myNet.neuPop[pop].isSigmaThres=True;
        myNet.neuPop[pop].sigmaThres=0.;
        myNet.neuPop[pop].muExt0=0.;
    DeltaT=100.;
    myNet.decod[0].tauD=50.;#lambda_D=20 Hz
    c=array([np.concatenate((linspace(0,0,runTime/(2.*DeltaT)),linspace(0,0,runTime/(2.*DeltaT)))) for com in xrange(M)]);
    for i in xrange(M):
        if myNet.neuPop[0].N<10:
            c[i,:]=0.5+0.*(-0.5+scipy.stats.uniform.rvs(size=1));
        else : 
            c[i,:]=20.+0.*(-0.5+scipy.stats.uniform.rvs(size=1));
elif typeExp[i]=="steps":
    if typeA=="leak":
        for pop in xrange(nPop):
            myRate.ratePop[pop].sigmaV=0.;
            myNet.neuPop[pop].sigmaV=0.;
            myNet.neuPop[pop].isSigmaThres=False;
            myNet.neuPop[pop].sigmaThres=0.;
            myNet.neuPop[pop].muExt0=0.;
        DeltaT=100.*dt;
        myNet.decod[0].tauD=50.; #lambda_D=20 Hz
        c0=0.3;
        c=c0+array([np.concatenate((linspace(0,0,runTime/(2.*DeltaT)),linspace(0,0,runTime/(2.*DeltaT)))) for com in xrange(M)]);
        for i in xrange(M):
            c[i,0.2*(runTime/(DeltaT)):0.4*(runTime/(DeltaT))]=c0+0.5;
            c[i,0.6*(runTime/(DeltaT)):0.8*(runTime/(DeltaT))]=c0-0.2;
    if typeA=="integrator":
        for pop in xrange(nPop):
            myRate.ratePop[pop].sigmaV=0.;
            myNet.neuPop[pop].sigmaV=0.;
            myNet.neuPop[pop].isSigmaThres=False;
            myNet.neuPop[pop].sigmaThres=0.;
            myNet.neuPop[pop].muExt0=0.;
        DeltaT=100.*dt;
        myNet.decod[0].tauD=50.; #lambda_D=20 Hz
        c0=0.;
        c=array([np.concatenate((linspace(0,0,runTime/(2.*DeltaT)),linspace(0,0,runTime/(2.*DeltaT)))) for com in xrange(M)]);
        for i in xrange(M):
            c[i,0.2*(runTime/(DeltaT)):0.4*(runTime/(DeltaT))]=c0+0.06;
            c[i,0.6*(runTime/(DeltaT)):0.8*(runTime/(DeltaT))]=c0-0.02;
elif typeExp[i]=="oscillating":
    for pop in xrange(nPop):
        myRate.ratePop[pop].sigmaV=0;
        myNet.neuPop[pop].sigmaV=0.01;
        myNet.neuPop[pop].isSigmaThres=False;
        myNet.neuPop[pop].sigmaThres=0.;
        myNet.neuPop[pop].muExt0=0.;
    DeltaT=dt;myRate.set_currentClock(_currClock=DeltaT);myNet.set_currentClock(_currClock=DeltaT);
    myNet.decod[0].tauD=0.75;
    c=array([np.concatenate((linspace(0,0,runTime/(2.*DeltaT)),linspace(0,0,runTime/(2.*DeltaT)))) for com in xrange(M)]);
    for i in xrange(M):
        c[i,:]=2.+1.*np.sin(2.*pi*0.1*array(arange(runTime/(DeltaT))));
elif typeExp[i]=="variable":
    for pop in xrange(nPop):
        myRate.ratePop[pop].sigmaV=0;
        myNet.neuPop[pop].sigmaV=0.;
        myNet.neuPop[pop].isSigmaThres=False;
        myNet.neuPop[pop].sigmaThres=0.;
        myNet.neuPop[pop].muExt0=0.;
    DeltaT=0.1;
    myNet.decod[0].tauD=50;
    c=array([np.concatenate((linspace(0,0,(int)(runTime/(2.*DeltaT))),linspace(0,0,(int)(runTime/(2.*DeltaT))))) for com in xrange(M)]);
    for i in xrange(M):
        c[i,0]=0.;
        for k in xrange((int)(runTime/(1.*DeltaT))):
            c[i,k]=max(1.,c[i,k-1]+0.1*(-0.5+scipy.stats.uniform.rvs(size=1)));
        
        window_len=11;
        def smooth(x,sigma):
             """ gaussian window smoothing """
             # extending the data at beginning and at the end
             # to apply the window at the borders
             s = numpy.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
             w = numpy.exp(-arange(-len(x)/100,len(x)/100)**2/(2.*sigma**2))
             y = numpy.convolve(w/w.sum(),s,mode='valid')
             return y[5:len(y)-5]
        filtC=smooth(c[i,:]/max(c[i,:]),10.);
        c[i,:]=abs(np.concatenate((array([0. for u in xrange(-len(filtC)+(int)(runTime/(DeltaT)))]),filtC)));

#########################################
###SET THE ASSOCIATED RATE MODEL ######
myRate.set_currentClock(_currClock=DeltaT);
myRate.muExt_t=c;
if typeA=="leak":
    myRate.A=diag(array([-0.2 for i in xrange(M)]));
if typeA=="integrator":
    myRate.A=diag(array([0. for i in xrange(M)]));
    
#########################################
###SET THE NETWORK####################
myNet.set_currentClock(_currClock=DeltaT);
myNet.muExt_t=c;
myNet.neuPop[0].name="";myNet.decod[0].name="";

#########################################
####SET THE CONNEXIONS###################
if typeCon=="fromGamma":

    ######BUILD FIRST GAMMA##
    if typeStru=="dense":
        myNet.decod[0].Gamma=(np.zeros((myNet.M,myNet.neuPop[0].N),dtype=float))+1.+0.*reshape(scipy.stats.norm.rvs(0.1,size=myNet.M*myNet.neuPop[0].N),(myNet.M,myNet.neuPop[0].N));
    elif typeStru=="sparse":
        myNet.decod[0].Gamma=matClass.buildMat(_M=myNet.M,_N=myNet.neuPop[pop].N,_J=1.,_R=1.,_dist="bernouilli",_type="all");
        if myNet.neuPop[0].N==3:
            myNet.decod[0].Gamma=linalg.cholesky(array([[5.,1.,0.],[1.,5.,1.],[0.,1.,5.]])).T
        
    #######THEN OMEGAS#####
    #I to I
    myNet.fastCon.set_Connexion(_pre=0,_post=0,_whatStruc="topo",_args=[scipy.sparse.lil_matrix(matClass.filterMat(-np.dot(myNet.decod[0].Gamma.T,myNet.decod[0].Gamma),_thre=0.000001))]);
    myNet.slowCon.set_Connexion(_pre=0,_post=0,_whatStruc="topo",_args=[scipy.sparse.lil_matrix(matClass.filterMat(np.dot(np.dot(myNet.decod[0].Gamma.T,(myRate.A+(1./(myNet.decod[0].tauD))*diag([1. for i in xrange(myNet.M)]))),myNet.decod[0].Gamma),_thre=0.000001))]);

elif typeCon=="fromOmega":
    
    ######BUILD FIRST OMEGA FAST##
    myOmega=matClass.matClass(_id=0,_name="fast",_N=myNet.neuPop[0].N,_M=myNet.M,_pConJ=0.2,_pConR=1,_J=1.,_R=20.);
    myOmega.searchOmegaDefPos();    
    myNet.fastCon.set_Connexion(_pre=0,_post=0,_whatStruc="topo",_args=[scipy.sparse.lil_matrix(matClass.filterMat(-myOmega.Omega,_thre=0.000001))]);
    
    ######FIND APPROPRIATE GAMMA#
    rot=matClass.buildMat(_M=myNet.M,_N=myNet.neuPop[0].N,_J=1.,_R=1.,_sdJ=0.5,_sdR=0.5,_dist="uniform",_isUni=True,_type="all");
    myNet.decod[0].Gamma=np.dot(rot,myOmega.OmegaChol);
    #myNet.decod[0].Gamma=myOmega.OmegaChol;
    #print(numpy.linalg.norm(myOmega.Omega-np.dot(myNet.decod[0].Gamma.T,myNet.decod[0].Gamma)))
    
    ######BUILD FINALLY OMEGA SLOW##
    myNet.slowCon.set_Connexion(_pre=0,_post=0,_whatStruc="topo",_args=[scipy.sparse.lil_matrix(matClass.filterMat(np.dot(np.dot(myNet.decod[0].Gamma.T,(myRate.A+(1./(myNet.decod[0].tauD))*diag([1. for i in xrange(myNet.M)]))),myNet.decod[0].Gamma),_thre=0.000001))]);

    
##########BUILD GAMMA THEN OMEGA################
myNet.neuPop[0].taum=20.;
if typeDelay=="off":
    myNet.fastCon.tauL[0,0]=0.;
    myNet.slowCon.tauL[0,0]=0.;
elif typeDelay=="on":
    myNet.fastCon.tauL[0,0]=1.;
    myNet.slowCon.tauL[0,0]=1.;
myNet.neuPop[0].V_Thres=array([myNet.neuPop[0].V_Rest[0]+0.*(max(myNet.nu,myNet.mu)/2.)+sum(myNet.decod[0].Gamma[:,i]**2)/2. for i in xrange(myNet.neuPop[0].N)]);
myNet.neuPop[0].V_Reset=myNet.neuPop[0].V_Thres-myNet.mu;

##########SET SOME THINGS TO RECORD################
myNet.maxVRecord=3;myNet.maxXRecord=3;
myNet.whoTraceX=[range(0,min(myNet.M,myNet.maxXRecord)) for pop in xrange(myNet.nPop)];
myNet.whoTraceV=[range(0,min(myNet.neuPop[pop].N,myNet.maxVRecord)) for pop in xrange(myNet.nPop)];

########################################################
#########BUILD TOOLS TOGETHER##########################
if myNet.checkDelayResetCompatibility() and myNet.checkGammaNoiseCompatibility():
    print("");
    print("CAN BUILD CONNECTION");
    myRate.build_Neu();myRate.build_Con();
    myNet.build_Neu();myNet.build_Con_NeuNeu();myNet.build_Decod();myNet.build_Con_NeuDecod();
    
```

```python
#########################################
###DO THE NEURONS NETWORK GRAPH####################
rc("font", family="serif", size=25)
rc("text", usetex=True)
rc("./graph.tex")

##############################
###INIT THE DAFT GRAPH#######
#geoType="layer";
geoType="circle";
if geoType=="layer":
    myNet.set_netGraph(width=20.,height=8.);
elif geoType=="circle":
    myNet.set_netGraph(width=20.,height=10.);

#################################
####SET PARAMETERS###############

###GLOBAL SIZE OF NODES#########
myNet.netGraph.sizeNode=2.5;
myNet.netGraph.sizeSpace=3.1;

###WHAT GEOMETRY FOR UNITS####
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].geometryType="layer";
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].geometryType=geoType;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].geometryType="layer";

###SPACE CONFIGURATION####
deltaL=2.;deltaLayer=2.;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_x=0.;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].C_x=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_x;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_x=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_x;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].deltaL=deltaL;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].deltaL=deltaL;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].deltaL=deltaL;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_y=3.;
if myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].geometryType=="layer":
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_y=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_y+deltaLayer;
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].C_y=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_y+deltaLayer;
elif myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].geometryType=="circle":
    for i in xrange(len(myNet.netGraph.popArray)):
        myNet.netGraph.popArray[i].theta0=np.pi;
        myNet.netGraph.popArray[i].theta=(max(1.,myNet.neuPop[0].N-1.))/2.;
        myNet.netGraph.popArray[i].thetaMax=np.pi;     
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].radius=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].deltaL*myNet.neuPop[0].N/2.;
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].deltaL=2.*myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].radius/(myNet.M-1.);
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].deltaL=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].deltaL;
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_x=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].deltaL*myNet.neuPop[0].N/2.;
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_y=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].C_y+deltaLayer;
    myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].C_y=myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].C_y+myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].radius+deltaLayer;
    
###HAVE UNITS DIFFERENT COLORS####
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].doDifferentFaceColor=True;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].doDifferentFaceColor=True;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].doDifferentFaceColor=True;

###DO WE WRITE NAME OF THE NODES ?##
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].noteNode=True;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].noteNode=True;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].noteNode=True;

###NUMBER OF SLOTS PER NODE####
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].nSlotsCon=16;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].nSlotsCon=2*16;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].nSlotsCon=16;

###AUTAPSE ANGLE####
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["command"])[0]].slotTheta=1;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["neuron"])[0]].slotTheta=1;
myNet.netGraph.popArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.popArray,["id"],["read"])[0]].slotTheta=1;

###DO WE DRAW CON ?###
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].drawCon=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].drawCon=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].drawCon=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].drawCon=False;

###DO WE HAVE TO DRAW NULL CONNECTIONS ?##
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].filterNullEdge=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].filterNullEdge=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].filterNullEdge=True;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].filterNullEdge=True;

##SIZE OF THE ARROW HEADS#####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].arrowHeadLength=0.75;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].arrowHeadLength=0.75;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].arrowHeadLength=1.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].arrowHeadLength=0.75;

##SIZE OF THE ARROWS WIDTHS#####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].arrowHeadWidth=0.1;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].arrowHeadWidth=0.1;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].arrowHeadWidth=0.1;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].arrowHeadWidth=0.25;

##STYLE OF THE ARROWS#####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].arrowStyle="wedge";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].arrowStyle="simple";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].arrowStyle="simple";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].arrowStyle="-[";

###CURVE OF THE EDGES####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].curveCon=0.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].curveCon=0.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].curveCon=0.2;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].curveCon=0.;

###AUTO CURVE OF THE EDGES####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].autoCurveCon=1.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].autoCurveCon=1.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].autoCurveCon=1.;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].autoCurveCon=1.;

###DO WE DRAW CONNEXIONS ?#####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].noteCon=False;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].noteCon=False;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].noteCon=False;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].noteCon=False;

####NAME OF THE CONNEXIONS####
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].name="";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].name="";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].name="";
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].name="";

###ASSOCIATE CON WITH MATRIX CONNECTIVITY
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["input"])[0]].conMat=myNet.decod[0].Gamma.T;
myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["output"])[0]].conMat=myNet.decod[0].Gamma;
if scipy.sparse.isspmatrix_lil(myNet.fastCon.Omega[0][0]):
    myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].conMat=myNet.fastCon.Omega[0][0].todense();
else:
    myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["fast"])[0]].conMat=myNet.fastCon.Omega[0][0];
if scipy.sparse.isspmatrix_lil(myNet.slowCon.Omega[0][0]):
    myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].conMat=myNet.slowCon.Omega[0][0].todense();
else:
    myNet.netGraph.conArray[toolsClass.findIndxObjFromAttr(myNet.netGraph.conArray,["id"],["slow"])[0]].conMat=myNet.slowCon.Omega[0][0];

#################################
####BUILD########################

####SET GEOMETRY######################
for i in xrange(len(myNet.netGraph.popArray)):
    myNet.netGraph.popArray[i].set_Geometry();

###DRAW NODES#######
for i in xrange(len(myNet.netGraph.popArray)):
    myNet.netGraph.popArray[i].draw();
 
###DRAW EGDES#######
for i in xrange(len(myNet.netGraph.conArray)):
    myNet.netGraph.conArray[i].draw();

######################################
#####RENDER THE GRAPH#################
if myNet.neuPop[0].N<10:
    myNet.netGraph.plotGraph();

```
